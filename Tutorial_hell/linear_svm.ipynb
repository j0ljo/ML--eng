```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.svm import SVC
```

```python
# 1. Create simple 2D dataset
X, y = make_blobs(n_samples=100, centers=2, random_state=5, cluster_std=1.2)
```

```python
# Train a linear SVM
model = SVC(kernel='linear')
model.fit(X, y)
```

```python
# Extract the model parameters
w = model.coef_[0]
b = model.intercept_[0]
```

```python
# Plot the data and decision boundary
plt.scatter(X[:,0], X[:,1], c=y, cmap='bwr')
```

```python
# Plot the data and decision boundary

x_line = np.linspace(X[:,0].min(), X[:,0].max(), 100)
y_line = -(w[0] * x_line + b) / w[1]
plt.plot(x_line, y_line, label = "Decision boundary")
```

```python
# Plot the margins
y_margin_up = -(w[0] * x_line + b - 1) / w[1]
y_margin_down = -(w[0] * x_line + b + 1) / w[1]
plt.plot(x_line, y_margin_up, 'k--', label = "Margins")
plt.plot(x_line, y_margin_down, 'k--')
```

```python
# Plot the support vectors
plt.scatter(model.support_vectors_[:,0],
            model.support_vectors_[:,1],
            s=120, edgecolors='k', facecolors='none',
            label='Support Vectors')
```

```python
plt.legend()
plt.title("Linear SVM with Decision Boundary and Margins")

plt.show()
```

```python
# Putting everything together

import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.svm import SVC

# 1. Create simple 2D dataset
X, y = make_blobs(n_samples=100, centers=2, random_state=5, cluster_std=1.2)

# 2. Train a Linear SVM
model = SVC(kernel='linear')
model.fit(X, y)

# 3. Extract model parameters
w = model.coef_[0]        # weight vector
b = model.intercept_[0]    # bias term

# 4. Plot
plt.scatter(X[:,0], X[:,1], c=y, cmap='bwr')

# Decision boundary: w·x + b = 0
x_line = np.linspace(min(X[:,0]), max(X[:,0]), 100)
y_line = -(w[0]/w[1]) * x_line - b/w[1]
plt.plot(x_line, y_line, label="Decision boundary")

# Margins: w·x + b = ±1
y_margin_up   = -(w[0]/w[1]) * x_line - (b-1)/w[1]
y_margin_down = -(w[0]/w[1]) * x_line - (b+1)/w[1]
plt.plot(x_line, y_margin_up,  '--', label="Margin")
plt.plot(x_line, y_margin_down,'--')

# Support vectors
plt.scatter(model.support_vectors_[:,0],
            model.support_vectors_[:,1],
            s=120, edgecolors='k', facecolors='none',
            label='Support Vectors')

plt.legend()
plt.title("Linear SVM Example")
plt.show()
```
